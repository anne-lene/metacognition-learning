```python
import sys
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
from utility_functions import *

plt.rcParams['mathtext.fontset'] = 'cm' 
np.set_printoptions(threshold=sys.maxsize)
font = {'size'   : 40,
        'family': 'arial'}
matplotlib.rc('font', **font)
plt.rcParams['figure.figsize'] = (15,15)
matplotlib.rcParams['axes.titlepad'] = 20
matplotlib.rcParams['axes.titlesize'] = 40
```

```python
# Constants and Variables
FEEDBACK_CONDITIONS = ['neg', 'neut', 'pos']
COLORS_BDI_GROUP = {1: 'green', 2: 'yellow', 3: 'orange', 4: 'red'}
COLORS_FEEDBACK={'pos': '#15B01A', 'neg': '#FF0000', 'neut': '#808080'}

DATA_PATH = './../data/processed/'
PROPS = {
    'boxprops':{'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}
```

```python
# Read data 
df_full = pd.read_csv(f'{DATA_PATH}variable_fb_data_full_processed.csv')
df_trials= pd.read_csv(f'{DATA_PATH}variable_fb_data_processed.csv')
df_sense=pd.read_csv(f'{DATA_PATH}variable_fb_data_metacognitive_analysis.csv')
df_alpha=pd.read_csv(f'{DATA_PATH}/variable_fb_data_processed_learning.csv')
df_pars=pd.read_csv(f'{DATA_PATH}/variable_fb_pars_fitted.csv')
df_first=df_trials[df_trials.trial==1]
df_first=standardise(df_first, 'confidence')
```

```python
def plot_violin_diff_perf(df):
    fig, ax = plt.subplots(figsize=(10, 10))
    make_figure(ax)
    sns.violinplot(data=df, x='correct', y='diff', color='lightgrey', linewidth=3, **PROPS)
    ax.scatter(range(3), df.groupby('correct').median()['diff'].values, c='blue', zorder=10, s=150, edgecolor='black')
    plt.xlabel('Difficulty', fontsize=60, labelpad=15)
    plt.ylabel('Estimation error', fontsize=60)
    plt.xticks(range(3), ['Easy', 'Medium', 'Hard'])
    plt.ylim([-90, 70])
    #plt.savefig('../figures/5_experimental_design_violin.jpg', dpi=300, bbox_inches='tight')
    plt.show()

    
def plot_histogram(df, variable, xlabel):
    fig, ax = plt.subplots(figsize=(10, 10))
    make_figure(ax)
    sns.histplot(df.groupby('pid').mean(), x=variable, bins=20, alpha=0.5, color='grey', kde=True,
                 line_kws={'linewidth': 5, 'color': 'black', 'alpha': 0.8}, edgecolor=None)
    ax.lines[0].set_color('black')
    plt.xticks(np.arange(0, 101, 20), np.arange(0, 101, 20))
    plt.xlabel(xlabel, fontsize=60)
    plt.ylabel('Frequency', fontsize=60)
    plt.xlim([20, 80] if variable == 'confidence' else [-0.9,0.3] )
    # plt.savefig(f'../figures/1_ef_experimental_design_hist_{variable}.jpg', dpi=300, bbox_inches='tight')
    plt.show()
    
    
def plot_mean_over_trials(var_dp, ylabel, ylimits):
    vals=df_trials.groupby(['trial'])[var_dp].agg(['mean', 'sem'])
    fb_colors = df_trials[['trial', 'condition']].drop_duplicates()['condition'].map(COLORS_FEEDBACK).values
    fig,ax=plt.subplots(figsize=(7,7))
    make_figure(ax)
    for i in range (14):
        ax.errorbar(range(i+1, i+3), vals['mean'][i:i+2], yerr=vals['sem'][i:i+2], 
                    c=fb_colors[i], linewidth=6, elinewidth=3, 
                     markersize=100)
    ax.bar(range(1,16), np.repeat(80, 15), alpha=0.1, color=fb_colors)
    ax.set_ylabel(ylabel)
    ax.set_xticks([1,5,10,15], [1,5,10,15])
    ax.set_xlabel('Trial')
    ax.set_xlim([0.5, 16])
    ax.set_ylim(ylimits)
    # plt.savefig('./../figures/conf_over_trials.jpg', dpi=300, bbox_inches='tight')
    plt.show()
    
    
def plot_bars():
    fig, ax = plt.subplots(figsize=(7,7))
    make_figure(ax)
    sns.barplot(data=df_trials, x="condition", 
                y='confidence', palette=COLORS_FEEDBACK, edgecolor='black', ax=ax)
    ax.set_xticklabels(['Neut', 'Pos', 'Neg'])
    plt.ylabel('SPE')
    plt.xlabel('')
    # plt.savefig('./../figures/conf_cond_bp.jpg', dpi=300, bbox_inches='tight')
    plt.show()


def plot_lm(dat, idp_var, dp_var, hue, xlabel, ylabel):
    sns.lmplot(data=dat, x=idp_var, y=dp_var, aspect=1, height=10, hue=hue,
           ci=95, scatter_kws={'s': 150, 'alpha': 0.4, 'edgecolor': 'black'}, line_kws={'lw': 6, 'alpha' : 1},
                palette=COLORS_FEEDBACK, legend=False, scatter=False)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # plt.savefig('./../figures/trial_level_update.jpg', dpi=300, bbox_inches='tight')
    plt.show()


def plot_metacog_measures(dat, var, label, color, alpha=1):
    dp_var='confidence_c' if var=='bias' else 'pid.diff_log_avg_c'
    ylim=[-1.5, 1] if var=='bias' else [-0.1, 0.4]  
    fig, ax = plt.subplots(figsize=(7,7))
    make_figure(ax)
    sns.regplot(data=dat, x='bdi_score', y=dp_var, color=color, ci=95, scatter_kws={'s': 100, 'alpha':0.3, 'edgecolor':'black'},
               line_kws={'alpha': alpha, 'linewidth':6}, scatter=False, label=label)
    plt.ylim(ylim)
    plt.xlabel('BDI score')
    plt.ylabel(f'Metacognitive {var}')
    plt.legend(fontsize=24)
    #plt.savefig('./../figures/metacog_sense.jpg', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def plot_model_comp():
    fig, ax = plt.subplots(1,3,figsize=(20,10), gridspec_kw={'width_ratios': [0.5, 2, 1]})
    [make_figure(ax[i]) for i in range(3)]
    ax[0].bar([1], df_pe['aic_diff'].mean(), yerr=[df_pe['aic_diff'].sem()], edgecolor='black', color='lightgrey', linewidth=2)
    for i, x in enumerate(['bdi_group_num', 'group_split_2']):
        sns.barplot(data=df_pe, x=x, y='aic_diff', errorbar= 'se', edgecolor='black', ax=ax[i+1], color='lightgrey', linewidth=2)
    for i, xlab in enumerate(['All', 'BDI group', 'Depressive group']):
        ax[i].set_xlabel(xlab)
    ax[0].set_xticklabels([''])
    ax[0].set_xticks([])
    ax[0].set_ylabel(r'AIC Difference, M2-M1') 
    ax[1].set_xticks(range(4), ['Minimal', 'Mild', 'Moderate', 'Severe'], fontsize=32)
    ax[2].set_xticks(range(0,2), ['Low', 'High'], fontsize=32)
    [ax[i].set_ylabel('') for i in range(1,3)]
    [ax[i].spines['left'].set_visible(False) for i in range(1,3)]
    [ax[i].set_yticks([]) for i in range(1,3)]
    [ax[i].set_ylim([0,3]) for i in range(3)]
    #plt.savefig(f'./../figures/modelcomparison_AIC.jpg', bbox_inches='tight', dpi=300)
    plt.show()

    
def plot_reg(ax, dp_var, col):
    sns.regplot(data=df_pars, x='bdi_score', y=dp_var, ax=ax, color=col, ci=95, line_kws={'lw': 6, 'alpha' : 1}, scatter=False)
    return 
    
    
def plot_pars(var, ylabel, ylim):
    fig, ax = plt.subplots(figsize=(10,10))
    make_figure(ax) 
    if var=='alpha': 
        for i, fb in enumerate(['pos', 'neut', 'neg']):
            plot_reg(ax, f'{var}_{fb}_rw_II', COLORS_FEEDBACK[fb])
        plot_reg(ax, f'{var}_rw_I', 'black')    
    else:
        plot_reg(ax, f'{var}_rw_I', 'black') 
        plot_reg(ax, f'{var}_rw_II', 'grey') 
    ax.set_ylabel(ylabel)
    ax.set_xlabel('BDI score') 
    plt.legend()
    plt.ylim(ylim)
    # plt.savefig('./../bdi_estimated_{var}.jpg', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def main():
    
    # Figure 5 Feedback modulates confidence and BDI score predicts negative metacognitive bias in a metacognitive task with variable feedback. 
    plot_histogram(df_trials, 'confidence', 'SPE')
    plot_histogram(df_trials, 'diff_log_avg', 'Performance')
    plot_violin_diff_perf(df_full)
    plot_mean_over_trials('confidence', 'SPE', [45, 75])   
    plot_mean_over_trials('diff_abs_avg', 'Performance', [0, 35])   
    plot_bars()
    plot_lm(df_trials, 'diff_abs_avg', 'confidence', 'condition_shifted', 'Absolute estmation error', 'SPE')
    plot_metacog_measures(df_first, 'bias', 'First trial', 'purple', 0.6)
    plot_metacog_measures(df_trials, 'bias', 'All trials', 'black', 0.6)
    plot_metacog_measures(df_sense, 'sensitivity', 'All trials', 'black', 0.6)


    # Figure 6 
    plot_lm(df_trials, 'bdi_score', 'confidence_update', 'condition', 'BDI score', 'Confidence update')
    plot_lm(df_trials[df_trials.trial==1], 'bdi_score', 'confidence', 'condition', 'BDI score', 'Empirical initial bias')
    plot_lm(df_alpha, 'bdi_score', 'alpha_empirical', 'condition', 'BDI score', r'Empirical learning rates, $\alpha_t$')
    plot_pars('bias', 'Estimated initial bias', [30, 80])
    plot_pars('alpha',  r'Estimated learning rates, $\hat{\alpha}_{+,0,-}$', [-0.1, 1])


if __name__ == '__main__':
    main()
    
```
